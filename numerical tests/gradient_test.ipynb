```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
from madam import Madam
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
torch.random.manual_seed(1)
np.random.seed(1)
n = 5
w = torch.rand(n)
x = torch.rand(n)
print(w)
class Loss(nn.Module):
    def __init__(self):
        super().__init__()
    def forward(self, x, mode):
        return torch.linalg.norm(x, mode)
```

```python
class Net(nn.Module):
    def __init__(self, w):
        super(Net, self).__init__()
        self.w = nn.Parameter(w)

    def forward(self, x):
        return self.w - x
```

```python
net = Net(w)
input_x = x
maxIter = 200
# optimizer = optim.Adam(net.parameters(), lr=0.01)
optimizer = Madam(net.parameters(), lr=0.01, p_scale=3.0, g_bound=10.0)
criterion = Loss()
```

```python
test = True
for i in range(maxIter):
    print(i, net.w)
    output = net(input_x)
    if test:
        if i < 100:
            loss = criterion(output, 8)
        else:
            loss = criterion(output, float('inf'))
    else:
        loss = criterion(output, 8)
    net.zero_grad()
    loss.backward()
    optimizer.step()
    print(net.w.grad)
    if(loss.item()<1e-4):
        print(i)
        break
print(loss.item())
```
